# Jewels and Stones

You're given strings J representing the types of stones that are jewels, and S representing the stones you have.  Each character in S is a type of stone you have.  You want to know how many of the stones you have are also jewels.

The letters in J are guaranteed distinct, and all characters in J and S are letters. Letters are case sensitive, so "a" is considered a different type of stone from "A".

## Example 1


Input: J = "aA", S = "aAAbbbb"
Output: 3

## Example 2:

Input: J = "z", S = "ZZ"
Output: 0

### Attempt 1

Firstly, I am going to approach the problem using Distionary and string. I am going to convert the S into dictionary in which the distinct variable will be added to the dictonary as key and the occurence of the string as value

```python
J="aA"
S="aAAbbbb"
stones = {}
```

```python
i=0
while(i<len(S)):
    if(S[i] not in stones):
         stones[S[i]]=1
    else:
        stones[S[i]]=stones[S[i]]+1
    print(S[i],stones)
    i+=1
```

```python
i=0
s = 0
while(i<len(J)):
    if(J[i] in stones):
        s = s+stones[J[i]]
    i+=1
```
